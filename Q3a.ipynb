```python
import numpy as np
import matplotlib.pyplot as plt
def both_ticks(fontsize):
    plt.tick_params(axis='both', which='major', direction="in", labelsize=fontsize*0.9, top=True, right=True)
    plt.tick_params(axis='both', which='minor', direction="in", labelsize=fontsize*0.9, top=True, right=True)

colours = [
    "#0077BB",  # Blue
    "#009988",  # Teal
    "#EE7733",  # Orange
    "#CC3311",  # Red
    "#EE3377",  # Magenta
]

def soln(T,x):
    return (np.sin(x*T)/x)**2

Ts = np.linspace(0.25, 1, 4)*2
print(Ts)
w=np.linspace(-10,10,1000)


fontsize = 8
plt.figure(figsize=[3*1.25,2*1.25], dpi=200)
for colour, T in zip(colours,Ts):
    plt.plot(w, soln(T,w), label =T, color=colour)

plt.rc('text', usetex=True)  # Enable LaTeX
plt.xlabel("$(\omega - \omega_0)/2$", fontsize=fontsize)
plt.ylabel("$|c_g(\omega,T)|^2$ (normalized)", fontsize=fontsize)
plt.rc('text', usetex=False)  # Disable LaTeX
both_ticks(fontsize)
plt.ylim(0,None)
plt.xlim(-8,8)
plt.legend(frameon=False, title="T values", title_fontsize=0.9*fontsize,fontsize=0.9*fontsize)
plt.tight_layout()
plt.show()
```
